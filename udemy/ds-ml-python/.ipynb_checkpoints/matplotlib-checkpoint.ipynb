```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
y = x*x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y, 'r*')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('X ao quadrado')
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])

axes1.plot(x, y)
axes2.plot(y, x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**3, 'b-*')
ax.plot(x, x**5, 'g--')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('graph')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
for axe in ax:
    axe.plot(x, y, 'b-*')
fig
#pode usar axe[0] pos ax é um array
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3)
ax[0][0].plot(x, y, 'r')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,4), dpi=100)
```

```python
fig, axes = plt.subplots(figsize=(5,5))
axes.plot(x, y, 'r-')
```

```python
fig.savefig('image.png')
```

```python
axes
```

```python
axes.plot(x, y, 'r--', label='x ^ 2')
axes.plot(y, x, 'g-*', label = 'sqrt(x)')
axes.legend()
```
